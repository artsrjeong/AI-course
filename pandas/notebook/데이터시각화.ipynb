```python
import seaborn as sns
import matplotlib.pyplot as plt

# 예제 데이터
tips = sns.load_dataset("tips")

# 성별(row)과 시간(col)에 따라 격자 생성
g = sns.FacetGrid(tips, row="sex", col="time", hue="smoker")

# 각 격자에 산점도 적용
g.map(sns.scatterplot, "total_bill", "tip")

plt.show()
```
